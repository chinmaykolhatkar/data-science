```python
import pandas as pd
from sklearn.model_selection import train_test_split

df = pd.read_csv('../input/melbourne-housing/melb_data.csv')
y = df.Price
X = df.drop(['Price'], axis=1)
X = X.select_dtypes(['int64', 'float64'])

train_x, validate_x, train_y, validate_y = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=1)
```

```python
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error

model = XGBRegressor()
model.fit(train_x, train_y)

mean_absolute_error(model.predict(validate_x), validate_y)
```

```python
model = XGBRegressor(n_estimators=500)
model.fit(train_x, train_y, early_stopping_rounds=5, eval_set=[(validate_x, validate_y)], verbose=True)

mean_absolute_error(model.predict(validate_x), validate_y)
```

```python
model = XGBRegressor(n_estimators=1000, learning_rate=0.05, n_job=4)
model.fit(train_x, train_y, early_stopping_rounds=5, eval_set=[(validate_x, validate_y)], verbose=True)

mean_absolute_error(model.predict(validate_x), validate_y)
```
